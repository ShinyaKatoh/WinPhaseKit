```python
of = open('station_for_REAL.dat', 'w')

with open('station_HAGI.list', 'r') as f:
    for line in f:
        if line.split()[4] == 'U':
            of.write('{:} {:} {:} {:} {:} {:}\n'.format(line.split()[14], 
                                                        line.split()[13], 
                                                        'CT', 
                                                        line.split()[3], 
                                                        'HHZ', 
                                                        float(line.split()[15])/1000 ))
of.close()
```
