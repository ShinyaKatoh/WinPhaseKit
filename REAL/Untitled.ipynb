```python
import numpy as np
import os
import shutil
```

```python
st = np.loadtxt('./station.dat', usecols=3, dtype='unicode')
st_lon = np.loadtxt('./station.dat', usecols=0)
st_lat = np.loadtxt('./station.dat', usecols=1)
st_dep = np.loadtxt('./station.dat', usecols=5)
```

```python
for lat in range(33,39):
    for lon in range(131,139):
        os.mkdir('./lat' + str(lat) + '-' + str(lat+1) + '_lon' + str(lon) + '-' + str(lon+1))
        fname = "station_lat" + str(lat) + "-" + str(lat+1) + "_lon" + str(lon) + "-" + str(lon+1) + ".dat"
        f = open(fname, 'w')
        for i in range(len(st)):
            if lat-1 <= st_lat[i] <= lat+2:
                if lon-1 <= st_lon[i] <= lon+2:
                    out = "{} {} {} {} {}\n".format(st_lon[i], st_lat[i], "WI", st[i], "HHZ", st_dep[i])
                    print(out)
                    f.write(out)
                    shutil.copy('/Users/shinyakato/Desktop/picks/for_REAL/WI.'+st[i]+'.P.txt', './lat' + str(lat) + '-' + str(lat+1) + '_lon' + str(lon) + '-' + str(lon+1) + '/WI.'+st[i]+'.P.txt')
                    shutil.copy('/Users/shinyakato/Desktop/picks/for_REAL/WI.'+st[i]+'.S.txt', './lat' + str(lat) + '-' + str(lat+1) + '_lon' + str(lon) + '-' + str(lon+1) + '/WI.'+st[i]+'.S.txt')
        f.close()
```
